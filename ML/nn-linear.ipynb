```python
diabetes = datasets.load_diabetes()
# Loading data for the x (we will only use one feature here)
diabetes_x = diabetes.data[:, 2]
print(diabetes_x.shape)
```

```python
diabetes_x = diabetes_x.reshape(-1,1) #converts to a 2 dimensional sckit learn
diabetes_x
```

### Linear Regression

```python
# Splitting training and testing data (features)
diabetes_x_train = diabetes_x[:-20]
diabetes_x_test = diabetes_x[-20:]

# Splitting training and testing data (target)
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]

# Linear regression
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(diabetes_x_train, diabetes_y_train)

# The coefficients
print('Coefficients:', regr.coef_)
# The mean square error
print("Residual sum of squares: %.2f" % np.mean((regr.predict(diabetes_x_test) - diabetes_y_test) ** 2))
# Explained variance score" 1 is perfect prediction
print('Variance score: %.2f' % regr.score(diabetes_x_test, diabetes_y_test))

# Plot outputs
plt.scatter(diabetes_x_test, diabetes_y_test, color='black')
plt.plot(diabetes_x_test, regr.predict(diabetes_x_test), color='blue', linewidth=3)
plt.show()
```

```python
# Create training data
x = np.array(list(range(0, 101))).reshape(-1, 1)
y = [val * 3 + 2 for val in x.flatten()]

# Linear regression model
regr = linear_model.LinearRegression()

# Create test data
x_train = np.array(list(range(101, 150))).reshape(-1, 1)
y_train = [val * 3 + 2 for val in x_train.flatten()]

# Train the model using the training sets
regr.fit(x, y)

# The coefficients
print('Coefficients:', regr.coef_)
print('Intecept', regr.intercept_)
# The mean square error
print("Residual sum of squares: %.2f" % np.mean((regr.predict(x_train) - y_train) ** 2))
# Explained variance score" 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x_train, y_train))

# Plot outputs
plt.scatter(x, y, color='black')
plt.plot(x_train, regr.predict(x_train), color='blue', linewidth=3)
plt.show()
```

### Knn Regression

```python
knn_regression = neighbors.KNeighborsRegressor(n_neighbors=4, weights='uniform')
knn_regression.fit(diabetes_x_train, diabetes_y_train)

# The mean square error
print("Residual sum of squares: %.2f" % np.mean((knn_regression.predict(diabetes_x_test) - diabetes_y_test) ** 2))
# Explained variance score" 1 is perfect prediction
print('Variance score: %.2f' % knn_regression.score(diabetes_x_test, diabetes_y_test))
# Plot outputs
plt.scatter(diabetes_x_test, diabetes_y_test, color='black')
plt.plot(diabetes_x_test, regr.predict(diabetes_x_test), color='blue', linewidth=3)
plt.scatter(diabetes_x_test, knn_regression.predict(diabetes_x_test), color='red', marker='x')
plt.show()
```

## Logistic Regression Model

```python
from sklearn import linear_model, datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics

iris = datasets.load_iris()
x = iris.data[:, :2] # we only take first two features
y = iris.target

logreg = linear_model.LogisticRegression(solver="lbfgs", multi_class="auto")
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

logreg.fit(x_train, y_train)
probs = logreg.predict_proba(x_test)

predicted = logreg.predict(x_test)

print("Confusion Set\n", metrics.confusion_matrix(y_test, predicted))

print(metrics.classification_report(y_test, predicted))
```
