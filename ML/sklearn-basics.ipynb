# Scikit-Learn Basics

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_theme()

import sklearn
from sklearn import linear_model
```

## Create and load data

```python
n = 200
m = 1.5 # slope
y = 13 # y-intercept
x = np.random.random(size=n) * 10 # Values are 0-10

# scikit learn expects data in column orientation
# All the same features should be the in the same column
x = x.reshape(-1,1) # converts row into column

ytrue = m * x + y
yrand = ytrue + np.random.normal(0, 3, size=n).reshape(-1, 1) #Adding Guassian error
```

```python
plt.scatter(x, ytrue)
plt.scatter(x, yrand)
```

```python
sklearn.metrics.mean_squared_error(ytrue, yrand)
```

## A Simple Model

```python
reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(x, yrand)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ypred = reg.predict(x)
plt.scatter(x, yrand)
plt.plot(x, ypred, color='r')
plt.plot(x, ytrue, color='g')
```

```python
sklearn.metrics.r2_score(ytrue, ypred)
```
